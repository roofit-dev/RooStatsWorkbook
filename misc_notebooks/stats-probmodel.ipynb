### Create a workspace with a Poisson probability model

```c++
using namespace RooFit;
```

```c++
RooWorkspace w("w") ;
w.factory("Poisson::p(N[0,20],mu[3])") ;
w.pdf("p")->SetTitle("Poisson") ;
```

### Create a canvas for two plots side by side

```c++
TCanvas* c = new TCanvas("c","c",1000,500) ;
c->Divide(2,1) ;
```

### Create the first plot

Make frame with 20 bins in range [0,20]

```c++
c->cd(1) ;
RooPlot* frame1 = w.var("N")->frame(0,20,20) ;
```

### Plot Poisson distribution in blue for $\mu=3$

```c++
w.var("mu")->setVal(3) ;
w.pdf("p")->plotOn(frame1,DrawOption("F"),FillColor(kBlue),Precision(1e-5),LineColor(kBlue)) ; 
```

### Adjust graphics and plot labels 

Draw plot frames on graphics pad

```c++
gPad->SetLeftMargin(0.15) ; 
frame1->SetTitle("#mu=3 (bkg only)") ;
frame1->GetXaxis()->SetTitle("Experimental outcome (event count)") ;
frame1->GetYaxis()->SetTitle("Poisson probability of outcome") ;
frame1->GetYaxis()->SetTitleOffset(1.8) ;

// Draw plot frame on graphics pad
frame1->Draw() ;
c->Draw();
```

### Create second plot

Make frame with 20 bins in range [0,20]

```c++
c->cd(2) ;
RooPlot* frame2 = w.var("N")->frame(0,20,20) ;
```

Plot Poisson distribution in red for mu=7 on frame 2

```c++
w.var("mu")->setVal(7) ;
w.pdf("p")->plotOn(frame2,DrawOption("F"),FillColor(kRed),Precision(1e-5),LineColor(kRed)) ;
```

### Adjust graphics and plot together

```c++
gPad->SetLeftMargin(0.15) ; 
frame2->SetTitle("#mu=7 (sig+bkg)") ;
frame2->GetXaxis()->SetTitle("Experimental outcome (event count)") ;
frame2->GetYaxis()->SetTitle("Poisson probability of outcome") ;
frame2->GetYaxis()->SetTitleOffset(1.8) ;
frame2->Draw() ;
c->Draw();
```
